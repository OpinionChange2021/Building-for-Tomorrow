## Sentiment BERT Unique

```python
for embedding in ['bert','roberta_']:
  for i, topic in enumerate(['Gender','sentiment_','Books']):

    Y1 = [2014,2013,2000]
    Y2 = [2020,2021,2019]

    bert_seasonal_aspects = pd.read_csv("/content/drive/MyDrive/" + embedding + "_common_aspects_similarities" +  topic + '.csv')[0:75]

    fig = plt.figure()
    ax = sns.boxplot(data=bert_seasonal_aspects[[str(x) for x in range(Y1[i],Y2[i])]], orient="h", palette="Set1")
    plt.savefig(topic + "_" + embedding + "_common.svg")
    fig.show()
    #fig.clf()
    print(len(bert_seasonal_aspects))
```
